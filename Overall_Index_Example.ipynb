```python
df1 = pd.read_csv("age_notation.csv")
print("DATASET HAS BEEN IMPORTED!")
```

Displaying the dataset

```python
del df1["country_code"]
del df1["source"]

df1.head()
```

```python
df1['new_cases']=df1['new_cases']/df1['population']
del df1['population']
df1.head()
```

```python
reshaped_df1=df1[df1['year_week']=='2021-43']
del reshaped_df1['year_week']

reshaped_df1=reshaped_df1.pivot(index='country', columns='age_group')
columns_name_ind=reshaped_df1.columns.tolist()

actual_columns=['country']

for item in columns_name_ind:
    actual_columns.append(item[0]+"_"+item[1])


new_df1=pd.DataFrame(columns=actual_columns,index=reshaped_df1.index)

i=0
for item in columns_name_ind:
    new_df1[actual_columns[i]]=reshaped_df1[item]
    i+=1
    
new_df1.head()
```

```python
for item in actual_columns:
    #Select Max_ratio
    Max_value=max(new_df1[item])
    #Tranform the column values to 0 to 100 index
    A=100/Max_value
    new_df1[item]=new_df1[item]*A

    
new_df1.head()
new_df1.to_csv("reshaped_age_notation.csv")
```

```python
#function to convert to alpah2 country codes and continents
from pycountry_convert import country_name_to_country_alpha3
def get_code(col):
    try:
        cn_a3_code =  country_name_to_country_alpha3(col)
    except:
        cn_a3_code = 'Unknown' 
    return cn_a3_code

overall_index['CODE']=None
for i in range(overall_index.shape[0]):
    overall_index['CODE'].iloc[i]=get_code(overall_index.iloc[i]['CountryName'])
overall_index.head()
```
